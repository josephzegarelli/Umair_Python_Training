```python
# make a function that takes name,age,city,extra and print details of him
```

```python
# positional, extra positoanl, keyword,extra
```

```python
def person(name , age, city="not declared",extra="human",*all_other_details,**kwarg): # parameters of function
    print(f"{name} lives in {city} his age is {age} and other thigns are {extra}")
    print(all_other_details)
    print(kwarg)
```

```python
person('joe' ,28,'Charlotte','hello') # apositional arguemnts
```

```python
person(28,'joe','oslo','learning pythong') # wrong positional arguemnt
```

```python
# solution
person(age=28,name='joe',city='city',extra='human') # keyword arguemnts
```

```python
# example of extra positonal arguennt
person('joe' ,28,'Charlotte','hello',1,2,3,4,5,study="Phd",height ="170") # apositional arguemnts
```
