```python
## Review, not in book
class DJ():
    def __init__(self, event, pay):
        self.event = event
        self.pay = int(pay)
    def view_booking(self):
        print(f'You are playing at {self.event} and you will be paid ${self.pay}')
```

```python
class Wedding(DJ):
    def __init__(self, couple, package):
        self.couple = couple
        event_name = (f"{self.couple.title()}'s wedding")
        self.package = package
        if self.package == 'a':
            pay_amount = '2000'
        elif self.package == 'b':
            pay_amount = '3000'
        else:
            print("Package code must be 'a' or 'b'")
        super().__init__(event_name, pay_amount)
    def view_wedding(self):
        print(f'You are playing at {self.event} and you will be paid ${self.pay}')
```

```python
class Wedding2(DJ):
    def __init__(self, couple, package,event, pay):
        super().__init__(event, pay)
        self.couple = couple
        self.package = package
        self.select_package()
    def select_package(self):
        if self.package == 'a':
            self.pay = 2000
        elif self.package == 'b':
            self.pay = 3000
        else:
            print("Package code must be 'a' or 'b'")
    def calculate_bill_based_on_hours(self, hours):
        hourly_rate = self.pay / 4  # assuming each package is for 4 hours
        total_bill = hourly_rate * hours
        return total_bill

    def tell_package_name(self):
        print(f'You have selected package {self.package} for {self.couple} wedding, and you will be paid ${self.pay}')
        
```

```python
class Bank_account():
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f'Deposited ${amount}. New balance is ${self.balance}.')
        else:
            print('Deposit amount must be positive.')

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f'Withdrew ${amount}. New balance is ${self.balance}.')
        else:
            print('Withdrawal amount must be positive and less than or equal to the balance.')

    def view_balance(self):
        print(f'Account holder: {self.account_holder}, Balance: ${self.balance}')
```

```python
class Synthesizer:
    def __init__(self, waveform, voices, detune):
        self.waveform = waveform
        self.voices = voices
        self.detune = detune
    def decrease_waveform(self):
        self.waveform -= 1
        print(f'Waveform changed to {self.waveform}')

    def add_voice(self):
        self.voices += 1
        print(f'Number of voices increased to {self.voices}')

    def set_detune(self, new_detune):
        self.detune = new_detune
        print(f'Detune set to {self.detune}')
    def play(self):
        print("Playing sound")

```

```python
s1 = Synthesizer("sawtooth", 8, 0.5)

s1.add_voice()
s1.set_detune(0.7)
```

```python
while True:
    s1.play()
    x = input("press1 to decrease waveform, press2 to add voice, press3 to set detune: press q to quit ")

    if x == '1':
        s1.decrease_waveform()
    elif x == '2':
        s1.add_voice()
    elif x == '3':
        new_detune = float(input("Enter new detune value: "))
        s1.set_detune(new_detune)
    elif x == 'q':
        break
    else:
        print("Invalid input. Please try again.")
```
