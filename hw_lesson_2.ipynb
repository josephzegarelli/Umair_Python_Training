```python
usernames = ['joe', 'joe2', 'joe3']
passwords = [1,2,3]
u = input('enter username: ')
if u in usernames:
    p = int(input('enter password: '))
    ind = usernames.index(u)
    if p == passwords[ind]:
        print('login succesful')
    else: 
        print('fail')
```
