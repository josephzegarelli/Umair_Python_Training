```python
class event():
    def __init__(self, name):
        self.name = name.title()
        self.costs = []
    def CreateGuestList(self):
        names = []
        name = input("Enter your name: ")
        while name.lower() != 'done':
            if name.title() == 'Sam':
                print('Sorry, Sam is not allowed')
                name = input("Enter your name: ")
            else:
                names.append(name.title())
                print(f'Guestlist: {names}')
                name = input("Enter your name: ")
        if name == 'done':
            print(f'Final Guestlist: {names}')
    def set_inital_cost(self, inital_cost):
        self.inital_cost = inital_cost
        self.costs.append(self.inital_cost)
    def add_cost(self, cost):
        self.cost = cost
        self.costs.append(self.cost)
        print(f'Total cost: {sum(self.costs)}')


```

```python
trio = event('trio')
trio.set_inital_cost(100)
trio.add_cost(50)
trio.CreateGuestList()
trio.add_cost(30)
```

```python
class Cat:
    def __init__(self,name,status="sleeping"):
        self.name = name
        self.status = status
    def meow(self):
        return "Meow!"

    def what_cats_doing(self):
        return f"{self.name} is {self.status}"
    def change_status(self,new_status):
        self.status = new_status
        
```
