```python
class Human():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def greeting(self):
        print(f'Hello I am {self.name.title()}, and I am {self.age} years old')
```

```python
my_human = Human('joe', 28)
```

```python
my_human.greeting()
```

```python
new_human = Human('Steve', 64)
new_human.greeting()
```
